```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fer_real = cv2.imread('ferrari.jpg')
fer_real = cv2.cvtColor(fer_real,cv2.COLOR_BGR2RGB)
plt.imshow(fer_real)
```

```python
fer_gray = cv2.cvtColor(fer_real,cv2.COLOR_BGR2GRAY)
plt.imshow(fer_gray,cmap='gray')
```

```python
gray = np.float32(fer_gray)
```

```python
gray
```

```python
#for ksize = 0.04

res = cv2.cornerHarris(gray,2,3,0.04)
res = cv2.dilate(res,None)
fer_real[res>0.01*res.max()] = [255,0,0]
plt.imshow(fer_real)
```

```python
#for ksize = 0.06

res = cv2.cornerHarris(gray,2,3,0.06)
res = cv2.dilate(res,None)
fer_real[res>0.01*res.max()] = [255,0,0]
plt.imshow(fer_real)
```
